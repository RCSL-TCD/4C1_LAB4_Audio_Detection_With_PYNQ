```python
import numpy as np
from pathlib import Path
OUT_NPZ = Path("lab_new/data/kws_12cls_mfcc_10x49_quant_flat.npz")
def inspect_npz(npz_path: Path):
    print(f"\n[INFO] checking: {npz_path}")
    data = np.load(npz_path, allow_pickle=True)

    X_train = data["X_train"]
    y_train = data["y_train"]
    X_val   = data["X_valid"]
    y_val   = data["y_valid"]
    X_test  = data["X_test"]
    y_test  = data["y_test"]
    label_names = data["label_names"]

    print("\n[INFO] label_names:", label_names)
    print("\n[INFO] shape checking:")
    print("  X_train:", X_train.shape, "y_train:", y_train.shape, "dtype:", X_train.dtype)
    print("  X_val  :", X_val.shape,   "y_val  :", y_val.shape,   "dtype:", X_val.dtype)
    print("  X_test :", X_test.shape,  "y_test :", y_test.shape,  "dtype:", X_test.dtype)
    print("  Y_train: ", y_train.shape, "dtype:", y_train.dtype)

    def print_label_stats(name, y):
        unique, counts = np.unique(y, return_counts=True)
        print(f"\n  {name} label distribution:")
        for u, c in zip(unique, counts):
            lbl = label_names[u] if u < len(label_names) else "NA"
            print(f"    idx={u:2d} ({lbl:8s}): {c:6d}")

    print_label_stats("Train", y_train)
    print_label_stats("Val",   y_val)
    print_label_stats("Test",  y_test)

inspect_npz(OUT_NPZ)
```

```python
import numpy as np
from pathlib import Path


IN_NPZ = Path("lab_new/data/kws_12cls_mfcc_10x49_quant_flat.npz")


OUT_DIR = Path("lab_new/data")
OUT_DIR.mkdir(parents=True, exist_ok=True)

data = np.load(IN_NPZ, allow_pickle=True)

X_test = data["X_test"]
y_test = data["y_test"]

np.save(OUT_DIR / "google_speech_X_test.npy", X_test)
np.save(OUT_DIR / "google_speech_y_test.npy", y_test)

print("[INFO] Saved:")
print("  google_speech_X_test.npy:", OUT_DIR / "google_speech_X_test.npy")
print("  google_speech_y_test.npy:", OUT_DIR / "google_speech_y_test.npy")
print("  Shapes:", X_test.shape, y_test.shape)
print("  Dtypes:", X_test.dtype, y_test.dtype)
```
